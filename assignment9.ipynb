```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 2*1000, endpoint=False)  # 1000 Hz sampling
x = np.cos(2*np.pi*10*t) * np.cos(2*np.pi*110*t)

plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('x(t)')
plt.title('Signal x(t) over 2 seconds')
plt.grid(True)
plt.show()
```

```python
Fs = 160
t = np.linspace(0, 2, 2*Fs, endpoint=False)
x = np.cos(2*np.pi*10*t) * np.cos(2*np.pi*110*t)
X = np.fft.fft(x)
f = np.fft.fftfreq(len(x), 1/Fs)

plt.plot(f[:len(f)//2], np.abs(X)[:len(f)//2])
plt.title('DFT magnitude (Fs=160 Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
Fs = 250
t = np.linspace(0, 40, 40*Fs, endpoint=False)
x = np.cos(2*np.pi*10*t) * np.cos(2*np.pi*110*t)
X = np.fft.fft(x)
f = np.fft.fftfreq(len(x), 1/Fs)

plt.plot(f[:len(f)//2], np.abs(X)[:len(f)//2])
plt.title('DFT magnitude (Fs=250 Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt

data = loadmat('handel.mat')
y = data['y'].squeeze()
Fs = int(data['Fs'])

Y = np.fft.fft(y)
f = np.fft.fftfreq(len(y), 1/Fs)

plt.plot(f[:len(f)//2], np.abs(Y)[:len(f)//2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Magnitude Spectrum of Audio')
plt.grid(True)
plt.show()
```

```python
power = np.abs(Y[:len(Y)//2])
top_indices = np.argsort(power)[-2:][::-1]
top_freqs = f[top_indices]
top_powers = power[top_indices]

print("Top frequencies:", top_freqs)
print("Corresponding powers:", top_powers)
```
